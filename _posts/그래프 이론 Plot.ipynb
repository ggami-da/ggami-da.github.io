```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a figure with two subplots for DAG1 and DAG2
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle("Directed Acyclic Graphs (DAGs)", fontsize=14)

# Subplot 1: DAG1 (A -> B)
G1 = nx.DiGraph()
G1.add_edges_from([("A", "B")])  # A -> B
pos1 = nx.spring_layout(G1)  # Layout to position the nodes
nx.draw(G1, pos1, with_labels=True, ax=axes[0], node_color='skyblue', edge_color='black', node_size=2500, arrowsize=20)
axes[0].set_title("DAG1: A -> B", fontsize=12)

# Subplot 2: DAG2 (B -> A)
G2 = nx.DiGraph()
G2.add_edges_from([("B", "A")])  # B -> A
pos2 = nx.spring_layout(G2)  # Layout to position the nodes
nx.draw(G2, pos2, with_labels=True, ax=axes[1], node_color='skyblue', edge_color='black',node_size=2500, arrowsize=20)
axes[1].set_title("DAG2: B -> A", fontsize=12)

# Adjust spacing and display the plots
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def draw_confounding_dag():
    G = nx.DiGraph()
    G.add_edges_from([
        ('X', 'A'),
        ('X', 'Y'),
        ('A', 'Y')
    ])

    pos = {'X': (0, 1), 'A': (1, 0), 'Y': (2, 1)}
    labels = {'X': 'X', 'A': 'A', 'Y': 'Y'}

    nx.draw(G, pos, with_labels=True, labels=labels, node_size=2000, node_color='lightblue', arrowsize=20)
    plt.title('Confounder DAG')
    plt.show()

draw_confounding_dag()
```

```python
def draw_frontdoor_direct():
    G = nx.DiGraph()
    G.add_edge('A', 'Y')

    pos = {'A': (0, 0), 'Y': (1, 0)}
    labels = {'A': 'A', 'Y': 'Y'}

    nx.draw(G, pos, with_labels=True, labels=labels, node_size=2000, node_color='lightblue', arrowsize=20)
    plt.title('A -> Y DAG')
    plt.show()

draw_frontdoor_direct()
```

```python
def draw_frontdoor_indirect():
    G = nx.DiGraph()
    G.add_edges_from([
        ('A', 'Z'),
        ('Z', 'Y')
    ])

    pos = {'A': (0, 0), 'Z': (1, 0), 'Y': (2, 0)}
    labels = {'A': 'A', 'Z': 'Z', 'Y': 'Y'}

    nx.draw(G, pos, with_labels=True, labels=labels, node_size=2000, node_color='lightblue', arrowsize=20)
    plt.title('A -> Z -> Y')
    plt.show()

draw_frontdoor_indirect()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def draw_confounding_dag():
    G = nx.DiGraph()
    G.add_edges_from([
        ('X', 'A'),
        ('X', 'Y'),
        ('A', 'Y')
    ])

    pos = {'X': (0, 1), 'A': (1, 0), 'Y': (2, 1)}
    labels = {'X': 'X', 'A': 'A', 'Y': 'Y'}

    plt.figure()
    nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=2000)
    nx.draw_networkx_labels(G, pos, labels, font_size=12)
    nx.draw_networkx_edges(G, pos, edgelist=G.edges(), arrows=True, arrowsize=20, node_size=2000)

    plt.title('Confounder DAG', fontsize=15)
    plt.axis('off')
    plt.show()

draw_confounding_dag()
```

```python
def draw_backdoor_blocked_dag():
    G = nx.DiGraph()
    G.add_edge('A', 'Y')
    G.add_node('X')

    pos = {'X': (0, 1), 'A': (1, 0), 'Y': (2, 1)}
    labels = {'X': 'X', 'A': 'A', 'Y': 'Y'}

    plt.figure()

    #  
    nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=1500)
    nx.draw_networkx_labels(G, pos, labels, font_size=12)

    # A -> Y  
    nx.draw_networkx_edges(G, pos, edgelist=[('A', 'Y')], arrows=True, arrowsize=20, node_size=1500)

    # X   'X'   
    plt.annotate('', xy=pos['A'], xytext=pos['X'],
                 arrowprops=dict(arrowstyle='->', linestyle='dashed', color='lightblue'))
    plt.annotate('', xy=pos['Y'], xytext=pos['X'],
                 arrowprops=dict(arrowstyle='->', linestyle='dashed', color='lightblue'))
    plt.text((pos['X'][0]+pos['A'][0])/2 - 0.1, (pos['X'][1]+pos['A'][1])/2 + 0.005, 'X', color='red', fontsize=20)
    plt.text((pos['X'][0]+pos['Y'][0])/2 - 0.1, (pos['X'][1]+pos['Y'][1])/2 + 0.005, 'X', color='red', fontsize=20)

    plt.title('Backdoor Path Blocking DAG', fontsize=15)
    plt.axis('off')
    plt.show()

draw_backdoor_blocked_dag()
```

```python
def draw_complex_confounding_dag():
    G = nx.DiGraph()
    G.add_edges_from([
        ('V', 'A'),
        ('V', 'W'),
        ('W', 'Y')
    ])

    pos = {'V': (0, 1), 'A': (1, 0), 'W': (1, 2), 'Y': (2, 1)}
    labels = {'V': 'V', 'A': 'A', 'W': 'W', 'Y': 'Y'}

    plt.figure()
    nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=2000)
    nx.draw_networkx_labels(G, pos, labels, font_size=12)

    nx.draw_networkx_edges(G, pos, arrows=True, arrowsize=20, node_size=2000)

    plt.title('DAG in complex confusion situations', fontsize=15)
    plt.axis('off')
    plt.show()

draw_complex_confounding_dag()
```

```python
def draw_frontdoor_with_mediator():
    G = nx.DiGraph()
    G.add_edges_from([
        ('A', 'Y'),
        ('A', 'Z'),
        ('Z', 'Y')
    ])

    pos = {'A': (0, 0), 'Z': (1, 1), 'Y': (2, 0)}
    labels = {'A': 'A', 'Z': 'Z', 'Y': 'Y'}

    plt.figure()
    nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=2000)
    nx.draw_networkx_labels(G, pos, labels, font_size=12)
    nx.draw_networkx_edges(G, pos, arrows=True, arrowsize=20, node_size=2000)

    plt.title('When A influences Y through Z and also has a direct influence', fontsize=15)
    plt.axis('off')
    plt.show()

draw_frontdoor_with_mediator()
```
